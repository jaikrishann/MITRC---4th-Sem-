```python
# One-time download script
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
df.to_csv("tips.csv", index=False)
```

```python
plt.plot(df["total_bill"])
```

```python
sns.lineplot(df["total_bill"])
```

```python
sns.scatterplot(x = df["total_bill"], y = df["tip"], data = df  )
plt.title("graph",fontsize = 20,
          color = "red",
          fontweight = "bold",
          backgroundcolor = "yellow",
          style = "italic")
```

```python
sns.boxplot(df["total_bill"])
```

```python
##outliers -- remove ya transform 
q1 = df["total_bill"].quantile(0.25)
q3 = df["total_bill"].quantile(0.75)
iqr = q3-q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
# df[df["total_bill"]>upper_bound]
# df[df["total_bill"]<lower_bound]
df = df[(df["total_bill"]>lower_bound) & (df["total_bill"]<upper_bound)]
sns.boxplot(df["total_bill"])
```

```python
sns.displot(df["total_bill"],kde =True)
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = df,kind="hex")
```
